```
# Step 1: Upload your input file
from google.colab import files

uploaded = files.upload()  # This will prompt you to upload your text file

# Step 2: Read the uploaded file and clean the text
input_filename = list(uploaded.keys())[0]  # Get the uploaded file name

cleaned_lines = []
with open(input_filename, 'r') as infile:
    for line in infile:
        # Strip leading/trailing spaces
        line = line.strip()
        # Skip blank lines
        if line:
            # Convert to lowercase
            cleaned_lines.append(line.lower())

# Step 3: Write cleaned text to a new file
output_filename = "clean_output.txt"
with open(output_filename, 'w') as outfile:
    outfile.write("\n".join(cleaned_lines))

print(f"Cleaned text has been saved to {output_filename}")

# Step 4: Download the cleaned file
files.download(output_filename)
```

```
import re

# Sample list of sentences
sentences = [
    "Hello, world! Welcome to Google Colab.",
    "This is python code; Python is powerful!",
    "Hello again: Python & Colab are amazing."
]

def preprocess_and_count(sentences):
    word_freq = {}

    for sentence in sentences:
        # Step 1: Remove special characters (keep only letters, numbers, and spaces)
        cleaned = re.sub(r'[^A-Za-z0-9\s]', '', sentence)

        # Step 2: Tokenize (split into words)
        tokens = cleaned.lower().split()

        # Step 3: Count frequency
        for word in tokens:
            word_freq[word] = word_freq.get(word, 0) + 1

    return word_freq

# Run the function
result = preprocess_and_count(sentences)

print("Final Word Frequency :")
print(result)
```
