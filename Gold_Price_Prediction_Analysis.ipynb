```
# =====================================================
# GOLD PRICE ANALYSIS USING PYTHON
# =====================================================
# IMPORT LIBRARIES
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================
# LOAD DATASET
# =====================================================

url = "https://raw.githubusercontent.com/sangitaGIT011/Myproj/refs/heads/main/gold_data_recent.csv"

df = pd.read_csv(url)

# =====================================================
# DISPLAY COLUMNS
# =====================================================

print(df.columns)

# =====================================================
# CREATE DATETIME COLUMN
# =====================================================

# Combine Date and Time columns

df['DateTime'] = pd.to_datetime(
    df['Date'].astype(str) + ' ' + df['Time'].astype(str),
    errors='coerce'
)

# Remove invalid rows

df = df.dropna(subset=['DateTime'])

# =====================================================
# CREATE ONLY DATE COLUMN
# =====================================================

df['Only_Date'] = df['DateTime'].dt.date

# =====================================================
# SORT DATA
# =====================================================

df = df.sort_values(by='DateTime')

# =====================================================
# DAILY GOLD PRICE
# =====================================================

daily_price = df.groupby('Only_Date')['Close'].mean()

# =====================================================
# CHANGE IN GOLD PRICE THROUGH DATE
# =====================================================

plt.figure(figsize=(18,8))

plt.plot(
    daily_price.index,
    daily_price.values,
    linewidth=2
)

plt.title('Change in Gold Price Through Date')

plt.xlabel('Date (Year-Month)')

plt.ylabel('Gold Closing Price (In $)')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()

# =====================================================
# TREND OF GOLD PRICE
# =====================================================

plt.figure(figsize=(18,8))

sns.lineplot(
    x=daily_price.index,
    y=daily_price.values
)

plt.title('Trend in Gold Price Through Date')

plt.xlabel('Date')

plt.ylabel('Gold Price')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()

# =====================================================
# DAILY PRICE CHANGE
# =====================================================

daily_change = daily_price.diff()

plt.figure(figsize=(18,8))

plt.plot(
    daily_change.index,
    daily_change.values
)

plt.title('Daily Change in Gold Price')

plt.xlabel('Date')

plt.ylabel('Price Change')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()

# =====================================================
# DISTRIBUTION OF DATA
# =====================================================

numeric_columns = df.select_dtypes(include=np.number).columns

df[numeric_columns].hist(
    figsize=(18,15),
    bins=30
)

plt.suptitle('Distribution of Data Across Columns')

plt.show()

# =====================================================
# BOXPLOT
# =====================================================

plt.figure(figsize=(18,10))

sns.boxplot(data=df[numeric_columns])

plt.title('Boxplot Distribution')

plt.xticks(rotation=45)

plt.show()

# =====================================================
# CORRELATION HEATMAP
# =====================================================

correlation = df[numeric_columns].corr()

plt.figure(figsize=(14,10))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Correlation Heatmap')

plt.show()

# =====================================================
# DATEWISE TABLE
# =====================================================

gold_table = pd.DataFrame({

    'Date': daily_price.index,

    'Gold Price': daily_price.values

})

print(gold_table.tail(20))

# =====================================================
# SAVE CSV
# =====================================================

gold_table.to_csv(
    'gold_price_datewise.csv',
    index=False
)

print("CSV File Saved Successfully")

# =====================================================
# END
# =====================================================
```

```
# =====================================================
# GOLD PRICE ANALYSIS
# =====================================================

# IMPORT LIBRARIES
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================
# LOAD DATASET
# =====================================================

url = "https://raw.githubusercontent.com/sangitaGIT011/Myproj/refs/heads/main/gold_data_recent.csv"

df = pd.read_csv(url)

# =====================================================
# DISPLAY AVAILABLE COLUMNS
# =====================================================

print("\nAvailable Columns:")
print(df.columns)

# =====================================================
# CREATE DATETIME COLUMN
# =====================================================

df['DateTime'] = pd.to_datetime(
    df['Date'].astype(str) + ' ' + df['Time'].astype(str),
    errors='coerce'
)

# REMOVE INVALID ROWS
# =====================================================

df = df.dropna(subset=['DateTime'])

# CREATE DATE COLUMN
# =====================================================

df['Only_Date'] = df['DateTime'].dt.date

# SORT DATA
# =====================================================

df = df.sort_values(by='DateTime')

# =====================================================
# USER INPUT SECTION
# =====================================================

print("\nEnter Date Range For Analysis")

start_date = input("Enter Start Date (YYYY-MM-DD): ")
end_date = input("Enter End Date (YYYY-MM-DD): ")

# CONVERT INPUT TO DATETIME
# =====================================================

start_date = pd.to_datetime(start_date)
end_date = pd.to_datetime(end_date)

# FILTER DATA BASED ON USER INPUT
# =====================================================

filtered_df = df[
    (df['DateTime'] >= start_date) &
    (df['DateTime'] <= end_date)
]

# CHECK DATA AVAILABILITY
# =====================================================

if filtered_df.empty:
    print("\nNo data found for entered date range.")

else:

    print("\nData Found Successfully")

    # =================================================
    # DAILY GOLD PRICE
    # =================================================

    daily_price = filtered_df.groupby('Only_Date')['Close'].mean()

    # =================================================
    # DISPLAY TABLE
    # =================================================

    gold_table = pd.DataFrame({

        'Date': daily_price.index,
        'Gold Price': daily_price.values

    })

    print("\nDatewise Gold Price")
    print(gold_table)

    # =================================================
    # LINE GRAPH
    # =================================================

    plt.figure(figsize=(18,8))

    plt.plot(
        daily_price.index,
        daily_price.values,
        linewidth=2
    )

    plt.title('Change in Gold Price Through Date')

    plt.xlabel('Date')

    plt.ylabel('Gold Closing Price')

    plt.xticks(rotation=45)

    plt.grid(True)

    plt.show()

    # =================================================
    # TREND GRAPH
    # =================================================

    plt.figure(figsize=(18,8))

    sns.lineplot(
        x=daily_price.index,
        y=daily_price.values
    )

    plt.title('Trend in Gold Price')

    plt.xlabel('Date')

    plt.ylabel('Gold Price')

    plt.xticks(rotation=45)

    plt.grid(True)

    plt.show()

    # =================================================
    # DAILY CHANGE
    # =================================================

    daily_change = daily_price.diff()

    plt.figure(figsize=(18,8))

    plt.plot(
        daily_change.index,
        daily_change.values
    )

    plt.title('Daily Change in Gold Price')

    plt.xlabel('Date')

    plt.ylabel('Price Change')

    plt.xticks(rotation=45)

    plt.grid(True)

    plt.show()

    # =================================================
    # HISTOGRAM
    # =================================================

    numeric_columns = filtered_df.select_dtypes(include=np.number).columns

    filtered_df[numeric_columns].hist(
        figsize=(18,15),
        bins=30
    )

    plt.suptitle('Distribution of Data Across Columns')

    plt.show()

    # =================================================
    # BOXPLOT
    # =================================================

    plt.figure(figsize=(18,10))

    sns.boxplot(data=filtered_df[numeric_columns])

    plt.title('Boxplot Distribution')

    plt.xticks(rotation=45)

    plt.show()

    # =================================================
    # HEATMAP
    # =================================================

    correlation = filtered_df[numeric_columns].corr()

    plt.figure(figsize=(14,10))

    sns.heatmap(
        correlation,
        annot=True,
        cmap='coolwarm',
        linewidths=0.5
    )

    plt.title('Correlation Heatmap')

    plt.show()

    # =================================================
    # SAVE CSV
    # =================================================

    file_name = input("\nEnter CSV File Name To Save: ")

    gold_table.to_csv(
        file_name + '.csv',
        index=False
    )

    print("\nCSV File Saved Successfully")
```
