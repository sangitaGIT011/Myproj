```
 #Practicle 7
#Data Cleaning and Preparation:
# a. Handle missing values and duplicates.
# b. Convert data types and normalize numerical columns.
# c. Rename and reorder dataset columns.

import pandas as pd
import seaborn as sns

data=sns.load_dataset('titanic')

# a. Handle missing values and duplicates.
##  Fill missing values

data['age']=data['age'].fillna(data['age'].median())
data['embarked']=data['embarked'].fillna(data['embarked'].mode()[0])

if 'Nspec' not in data['deck'].cat.categories:
  data['deck'] = data['deck'].cat.add_categories(['Nspec'])
data['deck'] = data['deck'].fillna('Nspec')

# b. Convert data types and normalize numerical columns.

# Convert 'Age' to integer (if itâ€™s float after filling NaN)
data['age'] = data['age'].astype(int)

# Normalize  'fare'
data['fare_normalized'] = (data['fare'] - data['fare'].min()) / (data['fare'].max() - data['fare'].min())



#print("\n--- Normalized Columns Preview ---")
print(data[['fare', 'fare_normalized']].head())


# c. Rename and reorder dataset columns.
# Rename selected columns for clarity
data.rename(columns={'sex': 'gender', 'embarked': 'embark_location','alive':'Alive','alone':'Alone'}, inplace=True)

#print(data.columns.tolist())

print(data.head())
## Apply new order (only include columns that exist)

## Move selected columns to front
cols_to_front = ['age']
new_columns = cols_to_front + [col for col in data.columns if col not in cols_to_front]
data = data[new_columns]

print(data.columns.tolist())




```

```
 #Practicle 9
#Data Visualization and EDA:
 #a. Plot histograms, bar charts, and line charts using matplotlib.
 #b. Visualize relationships with scatterplots and boxplots.
 #c. Generate correlation heatmaps and draw insights.

 # ======================================
# ðŸ”¹ IMPORT LIBRARIES
# ======================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ======================================
# ðŸ”¹ LOAD DATASET
# ======================================
data = sns.load_dataset('mpg')

print("\n--- Dataset Preview ---")
print(data.head())
print("\n--- Basic Info ---")
print(data.info())
print("\n--- Summary Statistics ---")
print(data.describe())

print("Histogram")
plt.figure(figsize=(6,4))
plt.hist(data['mpg'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Miles Per Gallon (MPG)')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

print("Bar Chart")
plt.figure(figsize=(6,4))
data.groupby('origin')['mpg'].mean().plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Average MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('Average MPG')
plt.show()

print("Line Chart")
plt.figure(figsize=(6,4))
data.groupby('model_year')['mpg'].mean().plot(kind='line', marker='o', color='orange')
plt.title('Average MPG by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.show()


 #b. Visualize relationships with scatterplots and boxplots.
print(" Relationships with scatterplots and boxplots.")

print('scatterplots')
plt.figure(figsize=(6,4))
sns.scatterplot(x='horsepower', y='mpg', hue='origin', data=data)
plt.title('Horsepower vs MPG by Origin')
plt.show()

print('Boxplots')
plt.figure(figsize=(6,4))
sns.boxplot(x='cylinders', y='mpg', data=data, color='skyblue')
plt.title('MPG Distribution by Cylinder Count')
plt.show()


print('correlation')

corr = data.select_dtypes(include='number').corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Heatmap of Vehicle Features')
plt.show()


```

```
#Practicle 11
#Working with Sets and Dictionaries:
 #a. Perform set operations: union, intersection, difference.
 #b. Add, update, and delete dictionary items.
 #c. Iterate through key-value pairs and check membership.


#a. Perform set operations: union, intersection, difference.

# Create two sets
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print("Set A:", set_a)
print("Set B:", set_b)

# Union â€“ elements from both sets
print("\nUnion:", set_a.union(set_b))

# Intersection â€“ common elements
print("Intersection:", set_a.intersection(set_b))

# Difference â€“ elements in A not in B
print("Difference (A - B):", set_a.difference(set_b))

# Symmetric Difference â€“ elements in either A or B, but not both
print("Symmetric Difference:", set_a.symmetric_difference(set_b))

# Membership Test
print("Is 3 in set A?", 3 in set_a)
print("Is 7 not in set A?", 7 not in set_a)

 #b. Add, update, and delete dictionary items.
#Create a dictionary
student = {
    "name": "Rohan",
    "age": 21,
    "course": "post graduation diploma in data science "
}
print("\nOriginal Dictionary:", student)

# Add new item
student["grade"] = "A"
print("After adding grade:", student)

# Update existing item
student["age"] = 22
print("After updating age:", student)

# Delete a specific key-value pair
del student["course"]
print("After deleting course:", student)

# Remove an item using pop()
removed = student.pop("grade")
print("Removed item:", removed)
print("After pop:", student)


#c. Iterate through key-value pairs and check membership.

# Rebuild dictionary
student = {"name": "Vishnu", "age": 23, "grade": "B"}

# Iterate through keys and values
print("\nIterating through dictionary:")
for key, value in student.items():
    print(f"{key}: {value}")

# Check membership
print("\nIs 'name' a key in student?", 'name' in student)
print("Is 'course' not a key in student?", 'course' not in student)

# Access keys and values separately
print("\nAll keys:", list(student.keys()))
print("All values:", list(student.values()))
```

```
#Practicle 13
 #Advanced Data Structure Applications:
 #a. Use nested lists and dictionaries for data storage.
 #b. Compare two lists for same elements (unordered).
 #c. Merge and sort dictionary data by key and value.

 # ==========================================
# Import required libraries
# ==========================================
import pandas as pd
import seaborn as sns

# Load a unique dataset
data = sns.load_dataset('penguins')
print(" Dataset Loaded Successfully!\n")
#print(data.head())

#a. Use nested lists and dictionaries for data storage.

# Select a few columns for structured data representation
subset = data[['species', 'island', 'bill_length_mm', 'bill_depth_mm']].dropna().head(5)

#  Option 1: Use unique keys (species + index)
nested_dict = {f"{row['species']}_{i}": {
                    'island': row['island'],
                    'bill_length_mm': row['bill_length_mm'],
                    'bill_depth_mm': row['bill_depth_mm']
                }
               for i, row in subset.head(5).iterrows()}

print("Nested Dictionary Example (Fixed):")
for key, value in nested_dict.items():
    print(f"{key} â†’ {value}")

    # Get species present on two islands
island_1 = data[data['island'] == 'Torgersen']['species'].dropna().unique().tolist()
island_2 = data[data['island'] == 'Dream']['species'].dropna().unique().tolist()

print("\nSpecies on Torgersen Island:", island_1)
print("Species on Dream Island:", island_2)

 #b. Compare two lists for same elements (unordered).
same_species = set(island_1) == set(island_2)
print("\nDo both islands have same species?", same_species)

# Create summary dictionaries
avg_bill_length = data.groupby('species')['bill_length_mm'].mean().round(2).to_dict()
avg_bill_depth  = data.groupby('species')['bill_depth_mm'].mean().round(2).to_dict()

print("\nAverage Bill Length per Species:", avg_bill_length)
print("Average Bill Depth per Species:", avg_bill_depth)

 #c. Merge and sort dictionary data by key and value.
merged = {sp: {'bill_length': avg_bill_length[sp], 'bill_depth': avg_bill_depth[sp]}
          for sp in avg_bill_length}

print("\n Merged Dictionary:")
for species, stats in merged.items():
    print(f"{species}: {stats}")

# Sort dictionary by species name (key)
sorted_by_key = dict(sorted(merged.items()))
print("\nSorted by Key (Species Name):", sorted_by_key)

# Sort dictionary by bill length (value)
sorted_by_length = dict(sorted(merged.items(), key=lambda x: x[1]['bill_length']))
print("\nSorted by Bill Length (Ascending):", sorted_by_length)


```
