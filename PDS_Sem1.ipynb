```
 #Practicle 7
#Data Cleaning and Preparation:
# a. Handle missing values and duplicates.
# b. Convert data types and normalize numerical columns.
# c. Rename and reorder dataset columns.

import pandas as pd
import seaborn as sns

data=sns.load_dataset('titanic')

# a. Handle missing values and duplicates.
##  Fill missing values

data['age']=data['age'].fillna(data['age'].median())
data['embarked']=data['embarked'].fillna(data['embarked'].mode()[0])

if 'Nspec' not in data['deck'].cat.categories:
  data['deck'] = data['deck'].cat.add_categories(['Nspec'])
data['deck'] = data['deck'].fillna('Nspec')

# b. Convert data types and normalize numerical columns.

# Convert 'Age' to integer (if it’s float after filling NaN)
data['age'] = data['age'].astype(int)

# Normalize  'fare'
data['fare_normalized'] = (data['fare'] - data['fare'].min()) / (data['fare'].max() - data['fare'].min())

#print("\n--- Normalized Columns Preview ---")
print(data[['fare', 'fare_normalized']].head())

# c. Rename and reorder dataset columns.
# Rename selected columns for clarity
data.rename(columns={'sex': 'gender', 'embarked': 'embark_location','alive':'Alive','alone':'Alone'}, inplace=True)

#print(data.columns.tolist())

print(data.head())
## Apply new order (only include columns that exist)

## Move selected columns to front
cols_to_front = ['age']
new_columns = cols_to_front + [col for col in data.columns if col not in cols_to_front]
data = data[new_columns]

print(data.columns.tolist())
```

```
#Practile 14
#Loop Applications and Recursion:
 #a. Calculate factorial of a number.
 #b. Generate Fibonacci series using recursion.
 #c. Compute GCD and LCM using recursion.

  #a. Calculate factorial of a number.
print('*Calculate factorial of a number*')
print('\n')

num=int(input('Enter a number: '))
def factorial(n):
    """Return factorial of n using recursion."""

    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(f"Factorial of {num} is: {factorial(num)}")

# b. Fibonacci series
print('\n')
print('Fibonacci Series')
num=int(input('Enter a number: '))
terms = int(input('Enter a terms: '))
def fibonacci(n):
    """Return nth Fibonacci number using recursion."""
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


print(f"\nFirst {terms} Fibonacci numbers:")
for i in range(terms):
    print(fibonacci(i), end=" ")
print()
 #c. GCD and LCM using recursion.
print('\n')
print('GCM AND LCM')
x=int(input('Enter a number: '))
y=int(input('Enter a number: '))

def gcd(a, b):
    """Return GCD using Euclidean recursion."""
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

def lcm(a, b):
    """Return LCM using GCD."""
    return abs(a * b) // gcd(a, b)

#x, y = 24, 36
print(f"\nGCD of {x} and {y} is: {gcd(x, y)}")
print(f"LCM of {x} and {y} is: {lcm(x, y)}")
```

```
#Practicle 18
#Scope and Recursion Concepts:
 #a. Demonstrate local, global, and nonlocal variables.
 #b. Use nested functions and variable scope.
 #c. Reverse string using recursion.

# a. Demonstrate local, global, and nonlocal variables

x = "Global Variable"  # Global variable

def outer_function():
    x = "Outer Variable"  # Enclosed (nonlocal) variable

    def inner_function():
        nonlocal x  # Refers to variable in outer_function
        x = "Modified by Inner Function"
        print("Inner Function:", x)

    inner_function()
    print("Outer Function:", x)

outer_function()
print("Global Scope:", x)

print("\n------------------------------")

# b. Use nested functions and variable scope

def calculator(a, b):
    def add():
        return a + b

    def multiply():
        return a * b

    def subtract():
        return a - b

    # Accessing inner functions (nested)
    print("Addition:", add())
    print("Multiplication:", multiply())
    print("Subtraction:", subtract())

calculator(10, 5)

print("\n------------------------------")

# c. Reverse string using recursion

def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

# Example
word = "Python"
print(f"Original String: {word}")
print(f"Reversed String: {reverse_string(word)}")
```

```
 #Practicle 25
 #Exception Handling Basics:
 #a. Implement try-except-finally blocks.
 #b. Handle ZeroDivisionError and FileNotFoundError.
 #c. Demonstrate nested exception handling.

# ----------------------------------------
# Exception Handling Basics
# ----------------------------------------

# a. Implement try-except-finally blocks
def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
    finally:
        print("Execution of divide_numbers() complete.\n")

# Example calls
divide_numbers(10, 2)
divide_numbers(5, 0)

# b. Handle ZeroDivisionError and FileNotFoundError
try:
    # Simulate a division error
    num1 = 8
    num2 = int(input("Enter a number to divide 8 by: "))
    print("Division result:", num1 / num2)

    # Try reading a non-existent file
    with open("non_existing_file.txt", "r") as f:
        content = f.read()
        print(content)

except ZeroDivisionError:
    print("You tried to divide by zero! Please enter a valid number.")
except FileNotFoundError:
    print("The file you are trying to open does not exist.")
except Exception as e:
    print("An unexpected error occurred:", e)
finally:
    print("Execution of File and Division operations completed.\n")

# c. Demonstrate nested exception handlin

def nested_exceptions():
    try:
        x = int(input("Enter an integer: "))
        try:
            result = 100 / x
            print("Inner Try: Division Successful. Result =", result)
        except ZeroDivisionError:
            print(" Division by zero not allowed.")
    except ValueError:
        print(" Invalid input! Please enter a number.")
    finally:
        print(" nested_exceptions() finished.\n")

nested_exceptions()
```

```
#Practicle 26
#Custom Exceptions and Validations:
 #a. Create user-defined exceptions.
 #b. Use assert statements for validation.
 #c. Handle multiple exceptions in one program

# a. Create user-defined exceptions

class NegativeValueError(Exception):
    """Raised when a negative value is entered."""
    pass

class AgeTooLowError(Exception):
    """Raised when age is below acceptable limit."""
    pass


def check_number(num):
    """Check for negative number using custom exception."""
    if num < 0:
        raise NegativeValueError("Negative numbers are not allowed.")
    else:
        print(f"Number {num} is valid.")


# ------------------------------------------------------------
# b. Use assert statements for validation
# ------------------------------------------------------------
def check_age(age):
    """Validate age using assert."""
    assert age >= 18, "Age must be at least 18 to proceed."
    print(f"Age {age} is valid for registration.")

# ------------------------------------------------------------
# c. Handle multiple exceptions in one program
# ------------------------------------------------------------
def main():
    try:
        num = int(input("Enter a number: "))
        check_number(num)

        age = int(input("Enter your age: "))
        check_age(age)

        # Additional operation that could raise multiple exceptions
        result = 100 / int(input("Enter a divisor: "))
        print(f"Division successful! Result = {result}")

    except NegativeValueError as nve:
        print("Custom Exception:", nve)
    except AssertionError as ae:
        print("Assertion Error:", ae)
    except ZeroDivisionError:
        print("Cannot divide by zero!")
    except ValueError:
        print("Invalid input! Please enter a number.")
    except Exception as e:
        print("Unexpected error:", e)
    else:
        print("All validations passed successfully!")
    finally:
        print("Program execution completed.")

main()
```

```
#Practicle 27
#Requests and Web Data Handling:
 #a. Download files using the requests module.
 #b. Parse and save web content to local files.
 #c. Handle connection and timeout exceptions

# Requests and Web Data Handling


import requests
from requests.exceptions import ConnectionError, Timeout, HTTPError
import os

# a. Download files using the requests module

def download_file(url, local_filename=None, timeout_seconds=10):
    try:
        # If no local name given, derive from URL
        if not local_filename:
            local_filename = url.split('/')[-1] or "downloaded_file"
        # Stream the download so large files are handled well
        with requests.get(url, stream=True, timeout=timeout_seconds) as r:
            r.raise_for_status()  # will raise HTTPError for bad HTTP status
            with open(local_filename, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
        print(f"Downloaded file saved as: {local_filename}")
        return local_filename
    except ConnectionError:
        print("Failed to connect to the URL:", url)
    except Timeout:
        print("The request timed out after", timeout_seconds, "seconds.")
    except HTTPError as http_err:
        print("HTTP error occurred:", http_err)
    except Exception as e:
        print("Some other error occurred:", e)
    return None

# b. Parse and save web content to local files
# ----------------------------------------
def fetch_and_save_webpage(url, local_filename, timeout_seconds=10):
    try:
        r = requests.get(url, timeout=timeout_seconds)
        r.raise_for_status()
        content = r.text  # use .text for text content (HTML etc)
        with open(local_filename, 'w', encoding=r.encoding or 'utf-8') as f:
            f.write(content)
        print(f"Webpage content from {url} saved to {local_filename}")
    except ConnectionError:
        print("Connection failed for:", url)
    except Timeout:
        print("Timeout when fetching:", url)
    except HTTPError as http_err:
        print("HTTP error occurred while fetching:", http_err)
    except Exception as e:
        print("Some other error occurred while fetching:", e)

# c. Handle connection and timeout exceptions
# ----------------------------------------
# (Handled inside the functions above)

# ----------------------------------------
# Example usage
# ----------------------------------------
if __name__ == "__main__":
    # Example a: Download a file
    file_url = "https://www.python.org/static/community_logos/python-logo-master-v3-TM.png"
    downloaded = download_file(file_url, "python_logo.png")

    # Example b: Fetch a web page and save
    page_url = "https://www.python.org/"
    fetch_and_save_webpage(page_url, "python_homepage.html")

    # Example: Attempt a bad URL to trigger exception handling
    bad_url = "https://nonexistent.python.org/404page"
    fetch_and_save_webpage(bad_url, "bad_page.html")

    # Example: Large download timeout test (you can simulate by using a slow or invalid URL)
    slow_url = "https://www.example.com/largefile.zip"
    _ = download_file(slow_url, "largefile.zip", timeout_seconds=2)

    print("\nCompleted Requests & Web Data Handling demo.")
```
