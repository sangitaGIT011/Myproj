```
#Practical 2
# Operators and Expressions:
 #a. Use arithmetic, comparison, and logical operators.
 #b. Perform string concatenation and type conversion.
 #2c. Evaluate mathematical expressions using precedence rules
# Use arithmetic, comparison, and logical operators.
# arithmetic operator
print("***Arithmetic operator***")
a=float(input( "1st number:"))
b=float(input( "2nd number:"))


print("Addition of Number:",a+b)
print("Subtraction of Number:",a-b)
print("Multiplication of Number:",a*b)
print("Division of Number:",a/b)

#Comparison operator
print("***Comparison operator***")
print("1st number is greater than 2nd number:",a>b)
print("1st number is equal to 2nd number:",a==b)
print("1st number is not equal to 2nd number",a!=b)

#logical operator
a=True
b=False
print("***logical operator***")
print("a and b:",a and b)
print("a or b:",a or b)
print("a not b:", not b)

#Perform string concatenation and type conversion
# String Concatenation
print("***String Concatenation***")
first_str = "python"
last_str = "programming"
full_str = first_str + " " + last_str
print("Full Name:", full_str)

# Type Conversion
print("***Type Conversion***")

num_str = "10"
num_int = int(num_str)  # Convert string to integer
sum_result = num_int + 25
print("Sum after type conversion:", sum_result)

#c. Evaluate mathematical expressions using precedence rules
# Operator Precedence Example
print("***Evaluate mathematical expressions***")
result = 12 + 6 * 2 ** 3 / 4 - 6
# According to precedence: Exponentiation > Multiplication/Division > Addition/Subtraction
print("Result of expression :", result)
```
